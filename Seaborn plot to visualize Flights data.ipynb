### Pivot this flights data such that it becomes a 2D matrix.

```python
flights_pv = flights.pivot_table(index='month', columns='year', values='passengers')
flights_pv.head()
```

# Heat map

```python
sns.heatmap(flights_pv)
```

# Cluster plot

```python
sns.clustermap(flights_pv)
```

```python
sns.jointplot( x ='year', y='passengers',  data=flights)
```

```python
sns.relplot(x='year', y= 'passengers', data = flights)
```

```python
sns.regplot(x='year', y= 'passengers', data = flights)
```

# Bar Plot

```python
sns.barplot(x = 'month', y = 'passengers', data = flights, ci= None)
```

# Pair Plot

```python
sns.pairplot(flights)
```
